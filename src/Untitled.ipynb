```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/yuyu/Project/VisClean/dataset/DBConf/DBPublications-input_id.csv')  
len(df)
```

```python
conf = ['VLDB', 'PVLDB','SIGMOD Conference', 'EDBT', 'ICDE', 'IEEE TKDE', 'CIDR']
```

```python
cnt = 0
for index, row in df.iterrows():
    for i in range(len(conf)):
        cit =  [0,0,0,0,0,0,0]
        if conf[i] in str(row['Venue']):
            if pd.isna(row['Citations']) == False:
                if row['Citations'] < 500:
                    cit[i] = row['Citations']
                    if cit[6] != 0 and cnt < 50:
                        cnt += 1
                        print("{\n type:'bar',\n stack:'stack',\n data:",cit,"\n},")
```
